```python
#Creating a dummy dataset

import numpy as np
import pandas as pd

labels = ['a','b','c']
my_data = [10,20,30]
arr = np.array(my_data)
d = {'a':10, 'b':20, 'c':30}
```

```python
#Printinf our data to check if it is created properly

print ('Labels are', labels)
print ('Data is', my_data)
print ('Dictionary is', d)
print('Array is', arr)
```

```python
#If you pass the list indise Seies it will create index by default

pd.Series(my_data)
```

```python
#If you need need customised indexes in a series than: 

pd.Series(my_data, index=labels )
```

```python
#Series of array

pd.Series(arr)
```

```python
#Indexing within the Series


print ('\nGrabbing the second element from the array\n', '-'*40,sep = '')
pd.Series(arr)[1]
```

```python
#Holding a text data in Series

print ('\n Holding a Text data\n', '-'*25, sep = '')
pd.Series(labels)
```

```python
#Creating dummy dataset

from numpy.random import randn as rn
matrix_data = rn(5,4)

row_labels = ['A','B','C','D','E']

columns_labels = ['W','X','Y','Z']

df = pd.DataFrame(matrix_data, row_labels, columns_labels)
df
```

```python
df.iloc[2] #To get the row index, we use iloc[]
```

```python
#Adding new column in the dataset

df['New (Sum of X and Z)'] = df['X'] + df['Z']
df
```

```python
#drop is used to drop a coulumn or row (axis = 0 is default axis for row where as axis = 1  is for colummn)

df.drop('E', inplace =True) #if inplace = True, the changes are permanent in the dataset

df.drop('X', inplace = True, axis=1) #Column dropping
df
```

```python
#Fetching row from the index label

df.loc['B']
```

```python
#Fetching row from the index position

df.iloc[0]
```
