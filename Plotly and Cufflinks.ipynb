```python
#Now import all the libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Now we need to use plotly and cufflinks in offline mode in your jupyter notebook

py.offline.init_notebook_mode(connected = True)
cf.go_offline()
```

```python
#Create a DatFrame
df = pd.DataFrame(np.random.randn(100,3), columns = ['A','B','C'])

#We will do some manipulation

df['A'] = df['A'].cumsum()+20
df['B'] = df['B'].cumsum()+20
df['C'] = df['C'].cumsum()+20

df.head()
```

```python
df.iplot()

#This will create a plotly graph which is interactive in nature
```

```python
df.plot()

#This will create a matplotlib graph which is static in nature
```

```python
df.iplot(x= 'A', y = 'B', mode = 'markers', size = 25)
```

```python
df.iplot(kind = 'bar', barmode = 'stack', bargap = 0.5)
```

```python
df.iplot(kind = 'barh', barmode = 'stack', bargap = 0.5) #Horizontal graph
```

```python
df.iplot(kind = 'box')

#Box plot divides the data in four parts 25% 50% 75% and 100%
#Range between q1 and q3 (actual box region) is called Inter Quartile Range (IQR)
#This graph is used maily to find outliers in the dataset which is represented by dots on top or below the box plot. Below graph does not have outliers
#Distance from minimum most line to q1  = q1 + (1.5 X IQR)
#Distance from q3 to maximum line  = q3 + (1.5 X IQR)

#Median line represents 0 Std. Deviation
#q3 represents 1 Std. Deviation 
#line between q3 and maximum represents 2 Std. Deviation
#Maximum represents 3 Std. Deviation
#Anything beyond maximum represents Maximum Outliers

#Median line represents 0 Std. Deviation
#q1 represents -1 Std. Deviation 
#line between q1 and minimum represents -2 Std. Deviation
#Minimum represents -3 Std. Deviation
#Anything lower than minimum represents Minimum Outliers
```

```python
df.iplot(kind = 'area', fill = True)
```
